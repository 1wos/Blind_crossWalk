```python
import json
import os
import cv2

def extract_labels(input_dir, specific_files):
    specific_files_labels = {}

    for filename in os.listdir(input_dir):
        if filename.endswith(".json"):
            json_path = os.path.join(input_dir, filename)
            with open(json_path, 'r', encoding='utf-8') as file:
                data = json.load(file)
                file_labels = []
                for shape in data['shapes']:
                    label = shape['label']
                    points = shape['points']
                    file_labels.append({
                        'label': label,
                        'points': points
                    })
                
                if filename.replace('.json', '') in specific_files:
                    specific_files_labels[filename] = file_labels

    return specific_files_labels

def draw_bounding_boxes(image_path, labels):
    image = cv2.imread(image_path)
    if image is None:
        print(f"이미지를 불러올 수 없습니다: {image_path}")
        return None
    for item in labels:
        label = item['label']
        points = item['points']
        # Assuming points are in [[x1, y1], [x2, y2]] format
        start_point = tuple(map(int, points[0]))
        end_point = tuple(map(int, points[1]))
        color = (0, 255, 0) if label == '1' else (255, 0, 0)
        cv2.rectangle(image, start_point, end_point, color, 2)
        # Draw the label above the bounding box
        text_position = (start_point[0], start_point[1] - 10)
        cv2.putText(image, label, text_position, cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
    return image

input_dir = r'D:\산학프로젝트용\(2차_최종) 교차로정보 데이터셋_20210720\교차로정보 데이터셋_bbox_2'
specific_files = ['MP_SEL_012921', 'MP_SEL_047407']
specific_files_labels = extract_labels(input_dir, specific_files)

for file, labels in specific_files_labels.items():
    print(f"{file}: {labels}")

    # Assuming the corresponding image file has the same name but with .jpg extension
    image_file = os.path.join(input_dir, file.replace('.json', '.jpg'))
    if os.path.exists(image_file):
        image_with_boxes = draw_bounding_boxes(image_file, labels)
        if image_with_boxes is not None:
            # Display the image
            cv2.imshow('Image with Bounding Boxes', image_with_boxes)
            cv2.waitKey(0)  # Wait for a key press to close the window
            cv2.destroyAllWindows()
    else:
        print(f"이미지 파일이 존재하지 않습니다: {image_file}")
```

```python
import json
import os

def json_to_yolo(input_dir, output_dir, class_mapping):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # 폴더 안의 첫 번째 JSON 파일만 변환
    for filename in os.listdir(input_dir):
        if filename.endswith(".json"):
            json_path = os.path.join(input_dir, filename)
            with open(json_path, 'r', encoding='utf-8') as file:
                data = json.load(file)
                yolo_labels = []
                img_width = data['imageWidth']
                img_height = data['imageHeight']
                for shape in data['shapes']:
                    label = shape['label']
                    points = shape['points']
                    class_id = class_mapping.get(label)

                    if class_id is None:
                        continue

                    # Convert points to YOLO format
                    x1, y1 = points[0]
                    x2, y2 = points[1]
                    x_center = ((x1 + x2) / 2.0) / img_width
                    y_center = ((y1 + y2) / 2.0) / img_height
                    width = (x2 - x1) / img_width
                    height = (y2 - y1) / img_height

                    yolo_labels.append(f"{class_id} {x_center} {y_center} {width} {height}")

                yolo_filename = os.path.join(output_dir, filename.replace('.json', '.txt'))
                with open(yolo_filename, 'w') as yolo_file:
                    yolo_file.write("\n".join(yolo_labels))

            # 첫 번째 파일만 처리하므로 루프 종료
            break

input_dir = r'D:\산학프로젝트용\(2차_최종) 교차로정보 데이터셋_20210720\교차로정보 데이터셋_bbox_6'
output_dir = os.path.join(input_dir, 'labels')
class_mapping = {
    'Zebra_Cross': 0,
    'R_Signal': 1,
    'G_Signal': 2,
    '1': 3,
}

json_to_yolo(input_dir, output_dir, class_mapping)
```

```python
import os
import shutil

def copy_to_yolo_directory(source_dir, dest_dir):
    if not os.path.exists(dest_dir):
        os.makedirs(dest_dir)

    # Copy imgs folder
    src_imgs = os.path.join(source_dir, 'imgs')
    dst_imgs = os.path.join(dest_dir, 'imgs')
    if os.path.exists(src_imgs):
        shutil.copytree(src_imgs, dst_imgs, dirs_exist_ok=True)

    # Copy labels folder
    src_labels = os.path.join(source_dir, 'labels')
    dst_labels = os.path.join(dest_dir, 'labels')
    if os.path.exists(src_labels):
        shutil.copytree(src_labels, dst_labels, dirs_exist_ok=True)

base_dir = r'D:\산학프로젝트용\(2차_최종) 교차로정보 데이터셋_20210720'
yolo_base_dir = r'D:\산학프로젝트용\교차로정보_yolo'

# 목록에 각 원본 폴더 추가
folders = ['교차로정보 데이터셋_bbox_2', '교차로정보 데이터셋_bbox_4', '교차로정보 데이터셋_bbox_6']

for folder in folders:
    source_dir = os.path.join(base_dir, folder)
    dest_dir = os.path.join(yolo_base_dir, folder)
    copy_to_yolo_directory(source_dir, dest_dir)
```
